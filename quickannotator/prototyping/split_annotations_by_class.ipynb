```python
import geojson

# Function to split GeoJSON by classification name
def split_geojson_by_name(input_geojson):
    # Load the GeoJSON data
    with open(input_geojson, 'r') as f:
        data = geojson.load(f)
    
    # Dictionary to store features by classification name
    split_data = {}
    
    # Iterate through features
    for feature in data:
        name = feature['properties'].get('classification', {}).get('name', 'Unknown')
        if name not in split_data:
            split_data[name] = {
                "type": "FeatureCollection",
                "features": []
            }
        split_data[name]["features"].append(feature)
    
    # Save each classification name to a separate GeoJSON file
    for name, geojson_data in split_data.items():
        output_file = f"{name.replace(' ', '_')}.geojson"
        with open(output_file, 'w') as f:
            geojson.dump(geojson_data, f)
        print(f"Saved {output_file}")

# Example usage
# split_geojson_by_name('input.geojson')
```
